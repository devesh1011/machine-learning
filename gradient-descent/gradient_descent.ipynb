```python
from sklearn.datasets import make_regression
import numpy as np

X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=10, random_state=1)

import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color="red")
```

```python
y_pred = ((28.39 * X) + 0).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred, color="blue", label="b = 0")
plt.legend()
plt.show()
```

```python
m = 28.39
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b  - step_size
b
```

```python
y_pred1 = ((28.39 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 28.39
lr = 0.01

epochs = 1000

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
